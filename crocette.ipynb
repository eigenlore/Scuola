```python
domande = [
    {
        "testo": "Sia $f:A \\rightarrow B$ una funzione iniettiva. Allora:",
        "risposte": [["Per ogni $x$ e $y$ in $A$, $f(x)\\neq f(y)$.", 1], ["Per ogni $x$ e $y$ in $A$, $f(x)= f(y)$.", 0], ["Esistono $x$ e $y$ in $A$ tali che $f(x)= f(y).$", 0], ["Per ogni $x$ e $y$ in $B$, $f(x)= f(y).$", 0]],
    },

    {
        "testo": "Sia $f:A \\rightarrow B$ una funzione suriettiva. Allora:",
        "risposte": [["Per ogni $y\\in B$ esiste $x \\in A$ tale che $y=f(x)$.", 1], ["Esiste $y\\in B$ tale che per ogni $x \\in A$ si ha che $y=f(x)$.", 0], ["Esiste $x\\in A$ tale che per ogni $y \\in B$ si ha che $y=f(x)$.", 0], ["Per ogni $x\\in A$ esiste $y \\in B$ tale che $x=f(y)$.", 0]],
    },

    {
        "testo": "Sia data la funzione $f:\\{0,1,2\\}\\rightarrow \\{4,5,6\\}$ definita da $f(0)=4,f(1)=5,f(2)=6$. Allora:",
        "risposte": [["La funzione è buinivoca.", 1], ["La funzione non è suriettiva.", 0], ["La funzione non è iniettiva.", 0], ["La funzione non è biunivoca.", 0]],
    },

    {
        "testo": "Sia $f: A \\rightarrow B$ una funzione. Si dice che $f$ è biunivoca (o biettiva) se:",
        "risposte": [["È sia iniettiva che suriettiva.", 1], ["È iniettiva ma non suriettiva.", 0], ["Il dominio A coincide con il codominio B.", 0], ["Esiste almeno un elemento di $B$ che non ha preimmagine.", 0]],
    },

    {
        "testo": "Data la funzione $f: \\mathbb{R} \\rightarrow \\mathbb{R}$ definita da $f(x) = x^2$, quale delle seguenti affermazioni è corretta?",
        "risposte": [["La funzione non è iniettiva perché $f(1) = f(-1)$.", 1], ["La funzione è suriettiva.", 0], ["La funzione è biunivoca.", 0], ["Per ogni $y \\in \\mathbb{R}$ esiste $x \\in \\mathbb{R}$ tale che $f(x) = y$.", 0]],
    },


    {
        "testo": "Sia $f: \\mathbb{R} \\rightarrow \\mathbb{R}$ definita da $f(x) = 2x + 3$. Quale proprietà soddisfa?",
        "risposte": [["La funzione è biunivoca.", 1], ["La funzione è solo iniettiva.", 0], ["La funzione è solo suriettiva.", 0], ["La funzione non è né iniettiva né suriettiva.", 0]],
    },

    {
        "testo": "Una funzione $f: A \\rightarrow B$ è suriettiva se e solo se l'insieme immagine $f(A)$ soddisfa:",
        "risposte": [["$f(A) = B$.", 1], ["$f(A) \\subseteq A$.", 0], ["$f(A)$ contiene un solo elemento.", 0], ["$f(A)$ è l'insieme vuoto.", 0]],
    },

    {
        "testo": "Sia $f: \\mathbb{Z} \\rightarrow \\mathbb{Z}$ definita da $f(n) = n + 1$. Essa è:",
        "risposte": [["Sia iniettiva che suriettiva.", 1], ["Iniettiva ma non suriettiva.", 0], ["Suriettiva ma non iniettiva.", 0], ["Né iniettiva né suriettiva.", 0]],
    },

    {
        "testo": "Cosa significa che una funzione $f: A \\rightarrow B$ NON è suriettiva?",
        "risposte": [["Esiste almeno un $y \\in B$ tale che per ogni $x \\in A, f(x) \\neq y$.", 1], ["Per ogni $y \\in B$ esiste un $x \\in A$ tale che $f(x) = y$.", 0], ["Esistono $x_1 \\neq x_2$ tali che $f(x_1) = f(x_2)$.", 0], ["Il dominio $A$ è un insieme vuoto.", 0]],
    },
    
    {
        "testo": "Qual è la definizione corretta di funzione tra due insiemi $A$ e $B$?",
        "risposte": [["Una relazione che associa ad ogni elemento di $A$ uno e un solo elemento di $B$.", 1], ["Una relazione che associa ad ogni elemento di $A$ almeno un elemento di $B$.", 0], ["Una relazione che associa ad ogni elemento di $B$ un unico elemento di $A$.", 0], ["Un insieme di coppie ordinate dove il primo elemento appartiene a $B$.", 0]],
    },

    {
        "testo": "Nel piano cartesiano, l'insieme dei punti che formano una funzione si chiama:",
        "risposte": [["Grafico della funzione.", 1], ["Immagine della funzione.", 0], ["Codominio della funzione.", 0], ["Relazione inversa.", 0]],
    },

    {
        "testo": "Data la funzione $f(x) = 3x - 1$, quanto vale l'immagine di $x = 2$?",
        "risposte": [["5.", 1], ["6.", 0], ["7.", 0], ["4.", 0]],
    },

    {
        "testo": "Come si può verificare graficamente se una curva nel piano cartesiano rappresenta una funzione?",
        "risposte": [["Ogni retta verticale deve intersecare la curva al massimo in un punto.", 1], ["Ogni retta orizzontale deve intersecare la curva in un solo punto.", 0], ["La curva deve passare per l'origine degli assi.", 0], ["La curva deve essere una retta.", 0]],
    },

    {
        "testo": "Se una funzione associa a ogni numero il suo doppio, la sua espressione analitica è:",
        "risposte": [["$f(x) = 2x$.", 1], ["$f(x) = x + 2$.", 0], ["$f(x) = x^2$.", 0], ["$f(x) = x/2$.", 0]],
    },

    {
        "testo": "L'insieme di tutti i valori assunti dalla $y$ nel codominio è detto:",
        "risposte": [["Insieme immagine.", 1], ["Dominio.", 0], ["Insieme di partenza.", 0], ["Zero della funzione.", 0]],
    },

    {
        "testo": "Sia $f: A \\rightarrow B$ una funzione e sia $y \\in B$. Cosa si intende per controimmagine di $y$?",
        "risposte": [["L'elemento (o gli elementi) $x \\in A$ tale che $f(x) = y$.", 1], ["Il valore $f(y)$ calcolato nel punto $y$.", 0], ["L'elemento $y$ moltiplicato per $-1$.", 0], ["Il valore che la funzione assume quando $x = 0$.", 0]],
    },

    {
        "testo": "Data la funzione $f(x) = x + 5$, qual è la controimmagine di $y = 12$?",
        "risposte": [["$x = 7$.", 1], ["$x = 17$.", 0], ["$x = 5$.", 0], ["$x = 12$.", 0]],
    },

    {
        "testo": "Nella funzione $f(x) = 2x$, qual è la controimmagine di $0$?",
        "risposte": [["$x = 0$.", 1], ["$x = 2$.", 0], ["Non esiste.", 0], ["Tutti i numeri reali.", 0]],
    },

    {
        "testo": "Se conosciamo il grafico di una funzione, come troviamo graficamente la controimmagine di un valore $y_0$?",
        "risposte": [["Troviamo le ascisse dei punti di intersezione tra il grafico e la retta orizzontale $y = y_0$.", 1], ["Troviamo l'ordinata del punto in cui la curva incrocia l'asse $y$.", 0], ["Troviamo il valore della funzione in $x = y_0$.", 0], ["Misuriamo la distanza dall'origine.", 0]],
    },

    {
        "testo": "Quali sono i tre concetti primitivi della geometria euclidea?",
        "risposte": [["Punto, retta e piano.", 1], ["Punto, segmento e angolo.", 0], ["Retta, piano e triangolo.", 0], ["Lunghezza, ampiezza e misura.", 0]],
    },

    {
        "testo": "Per due punti distinti del piano:",
        "risposte": [["Passa una e una sola retta.", 1], ["Passano infinite rette.", 0], ["Passano esattamente due rette.", 0], ["Non passa alcuna retta.", 0]],
    },

    {
        "testo": "Si definisce semiretta:",
        "risposte": [["La parte di retta costituita da un punto, detto origine, e da tutti i punti che lo seguono (o lo precedono).", 1], ["La metà esatta di una retta finita.", 0], ["L'insieme dei punti compresi tra due punti detti estremi.", 0], ["Una linea che ha due origini.", 0]],
    },

    {
        "testo": "Due segmenti si dicono adiacenti quando:",
        "risposte": [["Sono consecutivi e appartengono alla stessa retta.", 1], ["Hanno un estremo in comune.", 0], ["Hanno la stessa lunghezza.", 0], ["Non si intersecano mai.", 0]],
    },

    {
        "testo": "Un angolo si dice 'piatto' quando:",
        "risposte": [["I suoi lati sono due semirette opposte.", 1], ["I suoi lati sono semirette sovrapposte.", 0], ["Misura 90 gradi.", 0], ["È la metà di un angolo retto.", 0]],
    },

    {
        "testo": "Cosa si intende per 'bisettrice' di un angolo?",
        "risposte": [["La semiretta che partendo dal vertice divide l'angolo in due parti congruenti.", 1], ["Una retta che taglia i lati dell'angolo.", 0], ["Il punto medio dell'angolo.", 0], ["La somma di due angoli adiacenti.", 0]],
    },

    {
        "testo": "Due angoli si dicono 'complementari' se la loro somma è:",
        "risposte": [["Un angolo retto ($90^\\circ$).", 1], ["Un angolo piatto ($180^\\circ$).", 0], ["Un angolo giro ($360^\\circ$).", 0], ["Un angolo nullo.", 0]],
    },

    {
        "testo": "Due figure geometriche sono 'congruenti' quando:",
        "risposte": [["Esiste un movimento rigido che le sovrappone perfettamente.", 1], ["Hanno la stessa forma ma dimensioni diverse.", 0], ["Si trovano sullo stesso piano.", 0], ["Hanno lo stesso numero di vertici.", 0]],
    },

    {
        "testo": "Che cos'è una poligonale?",
        "risposte": [["Una figura formata da una sequenza ordinata di segmenti consecutivi.", 1], ["Una linea curva chiusa.", 0], ["L'insieme dei punti interni a un poligono.", 0], ["Una retta che attraversa più piani.", 0]],
    },

    {
        "testo": "Un poligono si dice 'convesso' se:",
        "risposte": [["Tutti i suoi angoli interni sono minori di un angolo piatto.", 1], ["Almeno un suo angolo interno è maggiore di un angolo piatto.", 0], ["Contiene i prolungamenti dei suoi lati.", 0], ["Ha tutti i lati uguali tra loro.", 0]],
    },

    {
        "testo": "Se un elemento del dominio ha 2 immagini distinte allora la funzione è",
        "risposte": [["Non è una funzione.", 1], ["Biunivoca.", 0], ["Iniettiva.", 0], ["Suriettiva.", 0]],
    }
   
]
n_alunni = 20
parola_segreta = "prova parallela classi prime"

import random
import base64


def sminchia(domande):
    random.shuffle(domande)
    for domanda in domande:
        random.shuffle(domanda["risposte"]) 

def inutile(i):
    if i==0:
        return 'a'
    elif i==1:
        return 'b'
    elif i==2:
        return 'c'
    elif i==3:
        return 'd'
    

def numb(letter): #letter a,b,c... are assigned 0,1,2... // space is assigned 26
    if letter ==' ':
        return 26
    else:
        return ord(letter)-ord('a')
def lett(number): #inverse of numb
    if number==26:
        return ' '
    else:
        return chr(ord('a')+number)
def shift(L,K): #shifts L by an amount given by numb(K). L can be a letter or a space.
    n_L=numb(L)
    n_K=numb(K)
    return lett((n_L+n_K)%27)
def shift_inv(L,K):
    n_L=numb(L)
    n_K=numb(K)
    return lett((n_L-n_K)%27)
def encryption(K, PT): #K key, PT plain text 
    while(len(K)<len(PT)):
        K+=K
    CT=''
    for i in range(len(PT)):
        CT+=shift(PT[i],K[i])
    return CT
def decryption(K,CT): #K key, CT cipher text
    while(len(K)<len(CT)):
        K+=K
    PT=''
    for i in range(len(CT)):
        PT+=shift_inv(CT[i],K[i])
    return PT

def cripta(domande):
    codice_verifica=""
    for domanda in domande:
        for i in range(len(domanda["risposte"])):
            if domanda["risposte"][i][1]==1:
                codice_verifica+=inutile(i)
    codice_verifica = encryption(parola_segreta, codice_verifica)
    return codice_verifica

def correggi(codice_verifica, risposte_studente):
    risposte_corrette = decryption(parola_segreta, codice_verifica)
    voto = 0
    for i in range(len(risposte_studente)):
        if risposte_corrette[i]==risposte_studente[i]:
            voto = voto + 1

    return risposte_corrette, voto ##formula Gozzano

def genera_output_latex(domande):

    codice_verifica=cripta(domande)

    output_latex = "\\textbf{Nome:}                                         \\textbf{Codice}: " 
    output_latex += codice_verifica
    output_latex += "\n\n\\begin{multicols}{2}\n\\begin{enumerate}\n"

    for domanda in domande:
        output_latex += "   \\item\n\n"+f" {domanda['testo']}\n"+"  \n\\begin{enumerate}[label=\Alph*.,nosep]\n"
        for risposta in domanda["risposte"]:
            output_latex += f"    \\item {risposta[0]}\n"
        output_latex += "  \\end{enumerate}\n\n"
    output_latex += "\\end{enumerate}\n\\end{multicols}\n\n\n\n\n\n\n\n\\newpage\n"
    
    return output_latex


# Genera le domande
for _ in range(n_alunni):
    sminchia(domande)
    print(genera_output_latex(domande))

```
