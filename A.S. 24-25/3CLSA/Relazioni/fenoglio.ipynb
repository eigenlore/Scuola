```python
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
import numpy as np

def lin(x, m, q):
    return x*m+q

#x=np.array([50,80,100,120])
#y=np.array([27.1,35,43,52])
#sigma_y=np.array([0.2,1.2,0.8,1.6])
x=np.array([100-6.5,120-6.5,60-6.5,80-6.5])
y=np.array([53.4-6.5,61-6.5,34.3-6.5,42.2-6.5])
sigma_y=np.array([1.4,2,0.9,0.8])
par, cov= curve_fit(lin, x, y, sigma=sigma_y, absolute_sigma=False)

xplot=np.linspace(x.min(), x.max(), num=100000)
plt.plot(xplot, par[0] * xplot + par[1], 'r', label='Risultato del fit')
plt.errorbar(x, y, yerr=sigma_y, fmt='o', label='Dati')
plt.xlabel(r'$h_0$')
plt.ylabel(r'$h_1$')
plt.title('Fit')
plt.legend()
plt.show()

coeff_ang=par[0]
terminenoto=par[1]
terminenoto_err=np.sqrt(cov[1][1])
print(coeff_ang)
print('Il termine noto è  =', terminenoto, '±', terminenoto_err)
```
