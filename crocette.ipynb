```python
domande = [
    {
        "testo": "Quale delle seguenti affermazioni è corretta riguardo alla pressione in un fluido in equilibrio? ",
        "risposte": [["La pressione è uguale in tutti i punti del fluido.", 1], [" La pressione è maggiore alla superficie del fluido rispetto alla profondità.", 0], ["La pressione aumenta con la profondità.", 0], ["La pressione dipende dalla velocità del fluido.", 0]],
    },
    {
        "testo": "La forza di Archimede:",
        "risposte": [["È sempre diretta verso l'alto.", 1], ["È sempre diretta verso il basso.", 0], ["È diretta verso l'alto se la densità del corpo è minore di quella del fluido.", 0], ["È diretta verso l'alto se la densità del corpo è maggiore di quella del fluido.", 0]],
    },
    {
        "testo": "Due corpi immersi hanno lo stesso volume $V$, ma forme e masse diverse. La forza di Archimede:",
        "risposte": [["È maggiore sul corpo con la massa maggiore.", 1], ["È maggiore sul corpo con la massa minore.", 0], ["È uguale sui due corpi.", 0], ["Dipende dalla forma del corpo.", 0]],
    },
    {
        "testo": "Una portata di 5 l/h (litri all'ora), corisponde in unità standard:",
        "risposte": [["$1,39\\cdot 10^6 \\textup{m}^3$/s.", 1], ["5 \\textup{m}^3$/s.", 0], ["5\\cdot 10^{-3} \\textup{m}^3$/s.", 0], ["1,39\\cdot 10^3 \\textup{m}^3$/s.", 0]],
    },
]
n_alunni = 3
parola_segreta = "private window firefox clears your search and browsing history when you close all private windows this doesnt make you anonymous"

import random
import base64


def sminchia(domande):
    random.shuffle(domande)
    for domanda in domande:
        random.shuffle(domanda["risposte"]) 

def inutile(i):
    if i==0:
        return 'a'
    elif i==1:
        return 'b'
    elif i==2:
        return 'c'
    elif i==3:
        return 'd'
    

def numb(letter): #letter a,b,c... are assigned 0,1,2... // space is assigned 26
    if letter ==' ':
        return 26
    else:
        return ord(letter)-ord('a')
def lett(number): #inverse of numb
    if number==26:
        return ' '
    else:
        return chr(ord('a')+number)
def shift(L,K): #shifts L by an amount given by numb(K). L can be a letter or a space.
    n_L=numb(L)
    n_K=numb(K)
    return lett((n_L+n_K)%27)
def shift_inv(L,K):
    n_L=numb(L)
    n_K=numb(K)
    return lett((n_L-n_K)%27)
def encryption(K, PT): #K key, PT plain text 
    while(len(K)<len(PT)):
        K+=K
    CT=''
    for i in range(len(PT)):
        CT+=shift(PT[i],K[i])
    return CT
def decryption(K,CT): #K key, CT cipher text
    while(len(K)<len(CT)):
        K+=K
    PT=''
    for i in range(len(CT)):
        PT+=shift_inv(CT[i],K[i])
    return PT

def cripta(domande):
    codice_verifica=""
    for domanda in domande:
        for i in range(len(domanda["risposte"])):
            if domanda["risposte"][i][1]==1:
                codice_verifica+=inutile(i)
    codice_verifica = encryption(parola_segreta, codice_verifica)
    return codice_verifica

def genera_output_latex(domande):
    #codice_verifica=cripta(domande)
    #print("\\textbf{Codice}:", codice_verifica)
    
    output_latex = "\\begin{center}  \\textbf{Griglia di valutazione}. \\begin{table}[h] \\centering\\begin{tabular}{|c|c|c|c|c|c|c|c|c|c|c|c|c|c|c|c|c|c|} \\hline Punti & $\\leq 6$& 7 & 8 & 9 & 10 & 11 & \\textbf{12} & 13 & 14 & 15 & 16 & 17 & 18 & 19 & 20  \\\\ \\hline Voto & 3 & 3.5 & 4 & 4.5 & 5 & 5.5 & \\textbf{6} & 6.5 & 7 & 7.5 & 8 & 8.5 & 9 & 9.5 & 10 \\\\ \\hline \\end{tabular} \\end{table} \\end{center}"
    output_latex += "\n\n\\begin{multicols}{2}\n\\begin{enumerate}\n"
    for domanda in domande:
        output_latex += "   \\item\n\n"+f" {domanda['testo']}\n"+"  \n\\begin{enumerate}[label=\Alph*.,nosep]\n"
        for risposta in domanda["risposte"]:
            output_latex += f"    \\item {risposta[0]}\n"
        output_latex += "  \\end{enumerate}\n\n"
    output_latex += "\\end{enumerate}\n\\end{multicols}\n\n\n\n\n\n\n\n\\newpage\n"
    
    return output_latex



def correggi(codice_verifica, risposte_studente):
    risposte_corrette = decryption(parola_segreta, codice_verifica)
    voto = 0
    for i in range(len(risposte_studente)):
        if risposte_corrette[i]==risposte_studente[i]:
            voto = voto + 1

    return risposte_corrette, voto/len(risposte_studente)*10

# Genera le domande
for _ in range(n_alunni):
    sminchia(domande)
    print(genera_output_latex(domande))

```
